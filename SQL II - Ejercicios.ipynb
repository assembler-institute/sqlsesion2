### 1.- The natural park

The Minister of the Environment has decided to create an information system on natural parks managed by each autonomous community. After a detailed analysis, the following conclusions have been reached:

- An Autonomous Community can have several Natural Parks. A Park may be shared by more than one Autonomous Community.

- A Natural Park is identified by a name, was declared on a certain date and is made up of several areas identified by a name and characterised by a certain extension.

- Each area is necessarily home to three types of species: plant, animal and mineral. Each species has a scientific name, a common name and an inventoried number of individuals per area. For the plant species we want to know if they flower and in what period this occurs; for the animal species we want to know their type of feeding and their mating period; for the mineral species we want to know if they are crystals or rocks.

- For the personnel working in each Natural Park, the DNI, NSS, name, address, telephone number and salary are kept. The following types of personnel are distinguished.

- Management personnel: They are stationed at an entrance located in the Natural Park.
    - Surveillance Staff: They watch over a specific area of the Park.
    - Conservation Staff: They maintain and conserve a specific area of the Park.
    - Research Staff: They have a qualification that has to be checked and can carry out research projects on a specific species.

- A Research Project has a code that identifies it, as well as a budget and a period of completion (f.ini f.fin).

### 2.- The Payroll system

A company decides to computerise its payroll. As a result of the analysis, the following information is obtained:

- Each employee is assigned a registration number when he/she joins the company, and this is the number used for internal identification purposes. In addition, the employee's NIF, name, number of children, withholding percentage for tax purposes, details of the current account into which the money is paid and the departments in which he/she works are recorded. 

- An employee can work in several departments. The name of a department is kept for each of its possible locations.

- Each employee is given multiple payslips throughout his working life in the company and at least one payslip per month.

- The total income received by the employee and the total deduction applied are the data of a pay slip. The distinction between two vouchers is made, in addition to the employee's registration number, we also want to know: the fiscal year and month number to which it belongs and an order number in the case of several vouchers received in the same month.

- Each payroll voucher consists of several lines and each line is identified by a line number of the corresponding voucher. A line may correspond to a payment or to a deduction. In both cases, the amount corresponding to the line is recorded; in the case of deductions, the basis on which the deduction is applied and the percentage applied for the calculation of the deduction are recorded.

- Each payment line of a payroll voucher corresponds to a single payment concept. In the same voucher, there may be several lines corresponding to the same pay concept. A code and a description are kept for each pay item.

- For company accounting purposes, each line of a voucher is allocated to at least one cost element. Several lines may be allocated to the same cost element. For each cost element, a code, a description and a balance are given.

- A hierarchy is established between the cost elements, in the sense that a cost element may contain other cost elements, but a cost element may only be contained in at most one other cost element.

### 3.- The non-profit association

The National Coordination of Non-Governmental Organisations wishes to maintain a database of the associations of this type that exist in our country. To do this, it needs to store information about each association, its members, the projects it carries out and its employees.

- For the associations, we want to store their CIF, name, address and province, their type, as well as whether they have been declared of public utility by the Ministry of the Interior.

- Each association is made up of members whose ID number, name, address, province, date of registration in the association, the monthly contribution they pay and the annual contribution they make are required.

- There are two types of workers in the association: employees and volunteers.

- Salaried employees are workers who receive a salary and hold a certain position in the association. We would like to record the amount they pay into the social security system and the amount of income tax deducted from them.

- The volunteers work in the organisation selflessly, and it is necessary to know their age, profession and the hours they dedicate to the association for the purpose of calculating statistics.

- Each worker is identified by his or her ID card, has a name and a date of entry.

- A member cannot be an employee of the association.

- The associations carry out projects. 

- For each project we want to store its identification number within the association, in which country it is carried out and in which area of the country, as well as the objective it pursues and the number of beneficiaries it affects. 

- A project is in turn made up of sub-projects. Sub-projects do not have other sub-projects.
